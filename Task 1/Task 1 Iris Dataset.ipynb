# Iris Dataset Visualization Project

## Introduction

This notebook performs basic data exploration and visualization using the famous Iris dataset.
We use Python libraries like Pandas, Seaborn, and Matplotlib to analyze patterns in the dataset.

## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Iris Dataset

We load the built-in Iris dataset from Seaborn.

```python
iris = sns.load_dataset("iris")
```

## Dataset Overview
We check the shape, columns, and first few rows of the dataset.

```python
print("Dataset Shape:")
print(iris.shape)

print("\nDataset Columns:")
print(iris.columns)

print("\nFirst 5 Rows:")
print(iris.head())
```

## Scatter Plot

This plot shows the relationship between sepal length and petal length for different iris species.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue="species")
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

## Histogram

This shows the distribution of sepal length in the dataset.

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=iris, x="sepal_length", bins=20, kde=True)
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

## Box Plot

This plot compares sepal length across different species.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=iris, x="species", y="sepal_length")
plt.title("Box Plot of Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length")
plt.show()
```

## Observation
The box plot compares sepal length across different iris species and helps identify variations and outliers.
